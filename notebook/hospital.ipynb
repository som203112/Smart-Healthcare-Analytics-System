Tkinter is Python's built-in library for creating Graphical User Interfaces (GUI). It provides a simple way to develop GUI applications using widgets like buttons, labels, text boxes, and more.

```python
#Creating a Simple Tkinter Window
from tkinter import *

# Create the main window
root = Tk()

# Set window title
root.title("My First Tkinter App")

# Set window size (width x height)
root.geometry("400x300")

# Run the application
root.mainloop()

#root.mainloop() keeps the window open and listens for user actions.
```

Tkinter provides several widgets to build interactive interfaces:

Widget	Description

Label	Displays text/images

Button	Creates clickable buttons

Entry	Input text box (for user input)

Text	Multi-line text input

Frame	Container for other widgets

Checkbutton	Checkbox for selection

Radiobutton	Allows selecting one option from multiple choices

Listbox	Displays a list of selectable items

Canvas	Draw shapes like lines, circles, and rectangles

Scrollbar	Adds scroll functionality


```python
from tkinter import *

def on_click():
    user_text = entry.get()## Get user input from the Entry widget....get() retrieves the text entered by the user
    label.config(text=f"Hello, {user_text}!")## Update the Label text
    #.config(text=...) updates the label's text dynamically.
##This function on_click() is a callback function that updates a Tkinter label with user input when triggered (e.g., by clicking a button).
root = Tk()
root.title("Simple Tkinter App")
root.geometry("300x200")

# Label
label = Label(root, text="Enter your name:")
label.pack()

# Entry box
entry = Entry(root)
entry.pack()

# Button
button = Button(root, text="Submit", command=on_click)
button.pack()

root.mainloop()
```
